```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt #import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
```

### Importing Data

```python
train_data = pd.read_csv(r'/Users/vidhanjain/desktop/Jupiter/digit-recognizer/train.csv')
test_data = pd.read_csv(r'/Users/vidhanjain/desktop/Jupiter/digit-recognizer/test.csv')
combine = [train_data,test_data]
train_data.shape
```

#### Data Split

```python
data_train = train_data.iloc[:31000]
data_dev = train_data.iloc[31000:]
```

```python
X_train = data_train.iloc[:,1:].to_numpy()
y_train = data_train[['label']].to_numpy()
X_dev = data_dev.iloc[:,1:].to_numpy()
y_dev = data_dev[['label']].to_numpy()
X_test = test_data.to_numpy()
```

### Building the Model

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dense(10)
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

### Training the Model

```python
model.fit(X_train, y_train, epochs=10)
```

```python
dev_loss, dev_acc = model.evaluate(X_dev,  y_dev, verbose=2)

print('\nTest accuracy:', dev_acc)
```

### Making Predictions

```python
probability_model = tf.keras.Sequential([model, 
                                         tf.keras.layers.Softmax()])
predictions = probability_model.predict(X_test)
```

```python
predictions[0]
```

```python
final_predictions = np.zeros([np.size(X_test,axis=0),1])
for i in range(0,np.size(X_test,axis=0)):
    final_predictions[i] = np.argmax(predictions[i])
```

```python
DF = pd.DataFrame(final_predictions)
DF.to_csv("submission.csv")
```
